```python
import csv
import json
import re

# def load_indexed_solutions(filepath):
#     solutions = {}
#     with open(filepath, 'r', encoding='utf-8') as f:
#         lines = [line.strip() for line in f if line.strip()]
#         for idx, solution in enumerate(lines, start=1):
#             solutions[str(idx)] = solution
#     return solutions

def load_human_solutions(filepath):
    solutions = {}
    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            parts = line.split()
            if len(parts) >= 2:
                problem_id = parts[0].replace('.', '')
                solution = parts[1]
                solutions[problem_id] = solution
    return solutions

# def load_problem_format_solutions(filepath):
#     solutions = {}
#     with open(filepath, 'r', encoding='utf-8') as f:
#         for line in f:
#             match = re.match(r"Problem\s+(\d+):\s*(\S+)", line.strip())
#             if match:
#                 problem_id, solution = match.groups()
#                 solutions[problem_id] = solution
#     return solutions

human_solutions   = load_human_solutions('Solutions.md')
# gpt_solutions     = load_indexed_solutions('GPT4.1.txt')
# gemini_solutions  = load_indexed_solutions('Gemini2.5Pro.txt')
# grok_solutions    = load_problem_format_solutions('Grok4.txt')
# claude_solutions  = load_problem_format_solutions('Claude4.0.txt')
# sonar_solutions   = load_indexed_solutions('Sonar.txt')
# gpt_solutions_2   = load_indexed_solutions('GPT4.1@2.txt')
# claude_solutions_2  = load_problem_format_solutions('Claude4.0@2.txt')

merged_data = []
unmatched_ids = []

with open('project_euler_problems.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        row = {key.strip(): value.strip() for key, value in row.items()}
        problem_id = row.get('id', '').strip()
        row['solution']             = human_solutions.get(problem_id)
        # row['gpt4_1_solution']      = gpt_solutions.get(problem_id)
        # row['gemini_2_5_solution']  = gemini_solutions.get(problem_id)
        # row['grok_4_solution']      = grok_solutions.get(problem_id)
        # row['claude_4_solution']    = claude_solutions.get(problem_id)
        # row['sonar_solution']    = sonar_solutions.get(problem_id)
        # row['gpt4_1_solution_2']       = gpt_solutions_2.get(problem_id)
        # row['claude_4_solution_2']       = claude_solutions_2.get(problem_id)
        if any(v is None for v in [
            row['solution']
            # ,
            # row['gpt4_1_solution'],
            # row['gemini_2_5_solution'],
            # row['grok_4_solution'],
            # row['claude_4_solution'],
            # row['sonar_solution'],
            # row['gpt4_1_solution_2'],
            # row['claude_4_solution_2']
        ]):
            unmatched_ids.append(problem_id)
        merged_data.append(row)

with open('dataset.json', 'w', encoding='utf-8') as outfile:
    json.dump(merged_data, outfile, indent=4, ensure_ascii=False)
```
